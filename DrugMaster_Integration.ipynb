Imports

```python
import pandas as pd
import json
import os
from datetime import datetime

pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", None)
```

Repository Locations

```python
OPENFDA_FOLDER = "repository"

RXNORM_FOLDER = "rxnorm_repository"

ORANGEBOOK_FOLDER = "orangebook_repository"

MASTER_FOLDER = "master_repository"
```

Create Master Folder

```python
if not os.path.exists(
    MASTER_FOLDER
):
    os.makedirs(
        MASTER_FOLDER
    )

print(
    os.path.abspath(
        MASTER_FOLDER
    )
)
```

Load OpenFDA Assets

```python
products_df = pd.read_parquet(
    f"{OPENFDA_FOLDER}/products.parquet"
)

packages_df = pd.read_parquet(
    f"{OPENFDA_FOLDER}/packages.parquet"
)

ingredients_df = pd.read_parquet(
    f"{OPENFDA_FOLDER}/ingredients.parquet"
)

repository_df = pd.read_parquet(
    f"{OPENFDA_FOLDER}/repository.parquet"
)
```

Load RxNorm Assets

```python
rxcui_lookup_df = pd.read_parquet(
    f"{RXNORM_FOLDER}/rxcui_lookup.parquet"
)

rxnorm_properties_df = pd.read_parquet(
    f"{RXNORM_FOLDER}/rxnorm_properties.parquet"
)

drug_repository_enriched_df = pd.read_parquet(
    f"{RXNORM_FOLDER}/drug_repository_enriched.parquet"
)
```

```python
print(
    f"Products: {len(products_df):,}"
)

print(
    f"Packages: {len(packages_df):,}"
)

print(
    f"Ingredients: {len(ingredients_df):,}"
)

print(
    f"Repository Rows: {len(repository_df):,}"
)

print(
    f"RxNorm Properties: {len(rxnorm_properties_df):,}"
)
```

Build Canonical Drug Master

```python
drug_master_df = (
    drug_repository_enriched_df
    .copy()
)

drug_master_df = (
    drug_master_df
    .drop_duplicates()
    .reset_index(drop=True)
)
```

Canonical Column Order

```python
master_columns = [

    "brand_name",

    "generic_name",

    "ingredient_name",

    "manufacturer",

    "product_ndc",

    "package_ndc",

    "package_description",

    "dosage_form",

    "route",

    "marketing_category",

    "application_number",

    "product_type",

    "rxcui",

    "name",

    "synonym",

    "tty"

]

available_columns = [

    col
    for col in master_columns
    if col in drug_master_df.columns
]

drug_master_df = (
    drug_master_df[
        available_columns
    ]
)
```

```python
print(
    f"Drug Master Rows: "
    f"{len(drug_master_df):,}"
)

display(
    drug_master_df.head(25)
)
```

```python
drug_master_df.to_parquet(

    f"{MASTER_FOLDER}/drug_master.parquet",

    index=False
)

print(
    "drug_master.parquet saved"
)
```

Save Metadata

```python
from datetime import datetime

master_metadata = {

    "build_timestamp":

        datetime.now().strftime(
            "%Y-%m-%d %H:%M:%S"
        ),

    "rows":

        int(
            len(drug_master_df)
        ),

    "unique_products":

        int(
            drug_master_df[
                "product_ndc"
            ].nunique()
        ),

    "unique_packages":

        int(
            drug_master_df[
                "package_ndc"
            ].nunique()
        ),

    "unique_rxcui":

        int(
            drug_master_df[
                "rxcui"
            ].nunique()
        )
}

master_metadata

drug_master_df.to_parquet(

    f"{MASTER_FOLDER}/drug_master.parquet",

    index=False
)

print(
    "drug_master.parquet saved"
)

with open(

    f"{MASTER_FOLDER}/drug_master_metadata.json",

    "w"
) as f:

    json.dump(
        master_metadata,
        f,
        indent=4
    )

print(
    "drug_master_metadata.json saved"
)
```

Build Data Dictionary

```python
drug_master_dictionary = pd.DataFrame({

    "column_name":

        drug_master_df.columns,

    "data_type":

        drug_master_df.dtypes.astype(str)
})

drug_master_dictionary.to_csv(

    f"{MASTER_FOLDER}/drug_master_data_dictionary.csv",

    index=False
)

print(
    "drug_master_data_dictionary.csv saved"
)
```

```python
ORANGEBOOK_FOLDER = "orangebook_repository"

orangebook_products_df = pd.read_parquet(
    f"{ORANGEBOOK_FOLDER}/orangebook_products.parquet"
)

orangebook_patent_df = pd.read_parquet(
    f"{ORANGEBOOK_FOLDER}/orangebook_patent.parquet"
)

orangebook_exclusivity_df = pd.read_parquet(
    f"{ORANGEBOOK_FOLDER}/orangebook_exclusivity.parquet"
)

print(
    f"Products: {len(orangebook_products_df):,}"
)

print(
    f"Patents: {len(orangebook_patent_df):,}"
)

print(
    f"Exclusivities: {len(orangebook_exclusivity_df):,}"
)
```

```python
drug_master_df[
    "application_number_join"
] = (

    drug_master_df[
        "application_number"
    ]

    .astype(str)

    .str.extract(
        r"(\d+)"
    )[0]

    .str.zfill(6)
)

orangebook_products_df[
    "application_number_join"
] = (

    orangebook_products_df[
        "application_number"
    ]

    .astype(str)

    .str.zfill(6)
)
```

```python
drug_master_v2_df = (

    drug_master_df.merge(

        orangebook_products_df[[
            "application_number_join",
            "application_type",
            "te_code",
            "approval_date",
            "RLD",
            "RS"
        ]],

        on="application_number_join",

        how="left"
    )
)

print(
    f"Rows: {len(drug_master_v2_df):,}"
)
```
