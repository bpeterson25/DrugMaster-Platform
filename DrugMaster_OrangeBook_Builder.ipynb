#  Load Orange Book Assets

```python
ORANGEBOOK_FOLDER = "orangebook_repository"

orangebook_products_df = pd.read_parquet(
    f"{ORANGEBOOK_FOLDER}/orangebook_products.parquet"
)

orangebook_patent_df = pd.read_parquet(
    f"{ORANGEBOOK_FOLDER}/orangebook_patent.parquet"
)

orangebook_exclusivity_df = pd.read_parquet(
    f"{ORANGEBOOK_FOLDER}/orangebook_exclusivity.parquet"
)
```

Standardize Appliation Number

```python
print("drug_master_df" in globals())

print("drug_master_v2_df" in globals())

print("repository_df" in globals())

print("drug_repository_enriched_df" in globals())
```

```python
drug_master_df[
    "application_number_join"
] = (

    drug_master_df[
        "application_number"
    ]

    .astype(str)

    .str.extract(
        r"(\d+)"
    )[0]

    .str.zfill(6)
)

orangebook_products_df[
    "application_number_join"
] = (

    orangebook_products_df[
        "application_number"
    ]

    .astype(str)

    .str.zfill(6)
)
```

Join Orange Book Product

```python
drug_master_v2_df = (

    drug_master_df.merge(

        orangebook_products_df[[
            "application_number_join",
            "application_type",
            "te_code",
            "approval_date",
            "RLD",
            "RS"
        ]],

        on="application_number_join",

        how="left"
    )
)
```

Patent Summary

```python
patent_summary_df = (

    orangebook_patent_df

    .groupby(
        "application_number"
    )

    .agg({

        "patent_number":
            "count"
    })

    .reset_index()

    .rename(

        columns={

            "application_number":
                "application_number_join",

            "patent_number":
                "patent_count"
        }
    )
)
```

Join Patent Counts

```python
drug_master_v2_df = (

    drug_master_v2_df.merge(

        patent_summary_df,

        on="application_number_join",

        how="left"
    )
)
```

Exclusivity Summary

```python
exclusivity_summary_df = (

    orangebook_exclusivity_df

    .groupby(
        "application_number"
    )

    .agg({

        "exclusivity_code":
            "count"
    })

    .reset_index()

    .rename(

        columns={

            "application_number":
                "application_number_join",

            "exclusivity_code":
                "exclusivity_count"
        }
    )
)
```

Join Exclusivity Counts

```python
drug_master_v2_df = (

    drug_master_v2_df.merge(

        exclusivity_summary_df,

        on="application_number_join",

        how="left"
    )
)
```
